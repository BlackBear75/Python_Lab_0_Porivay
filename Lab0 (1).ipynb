```python
def decimal_to_binary_integer_part(n):
    if n == 0:
        return "0"
    
    binary = ""
    while n > 0:
        binary = str(n % 2) + binary
        n //= 2
    return binary

def decimal_to_binary_fractional_part(fraction, precision=10):
    binary = ""
    count = 0
    
    while fraction > 0 and count < precision:
        fraction *= 2
        if fraction >= 1:
            binary += '1'
            fraction -= 1
        else:
            binary += '0'
        count += 1
    
    return binary

def decimal_to_binary(n, precision=10):
    if n < 0:
        return '-' + decimal_to_binary(-n, precision)
    
    integer_part = int(n)
    fractional_part = n - integer_part
    
    binary_integer_part = decimal_to_binary_integer_part(integer_part)
    
    if fractional_part > 0:
        binary_fractional_part = decimal_to_binary_fractional_part(fractional_part, precision)
        return f"{binary_integer_part}.{binary_fractional_part}"
    else:
        return binary_integer_part

print(decimal_to_binary(10))          # 1010
print(decimal_to_binary(10.25))       # 1010.01
print(decimal_to_binary(0.625))       # 0.101
print(decimal_to_binary(-7.75))       # -111.11
```
